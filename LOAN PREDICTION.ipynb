```python
mode = data['Married'].mode()
data['Married'].fillna(data['Married'].mode()[0], inplace=True)
```

```python
mode = data['Dependents'].mode()
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)
```

```python
data['Dependents'] = data['Dependents'].replace({'3+': 4}).astype(int)
```

```python
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
```

```python
data['Married'] = data['Married'].map({'Yes':0,'No':1})
```

```python
data['Education'] = data['Education'].map({'Not Graduate':0,'Graduate':1})
```

```python
data['Self_Employed'] = data['Self_Employed'].map({'No':0,'Yes':1})
```

```python
data['Loan_Status'] = data['Loan_Status'].map({'N':0,'Y':1})
```

```python
data['Property_Area'] = data['Property_Area'].map({'Urban':0,'Semiurban':1,'Rural':2})
```

```python
x = data.drop("Loan_Status",axis =1)
y = data['Loan_Status']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.49,random_state=1)
```

```python
reg_model = LogisticRegression()
reg_model.fit(x_train,y_train)
```

```python
reg_model.score(x_train,y_train)
```

```python
reg_model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfcl =  RandomForestClassifier(n_estimators=8,random_state=1)
rfcltr =  rfcl.fit(x_train,y_train)
rfclte =  rfcl.fit(x_test,y_test)
rfcltr.score(x_test,y_test)
```
